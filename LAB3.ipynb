```python
#code to clean up olsat verbal score


for i in range(117):
    if (type(data['OLSAT Verbal Score'][i]) == str):
        try:
            #print("value i is ",i,int(data['OLSAT Verbal Score'][i]))
            data['OLSAT Verbal Score'][i]= int(data['OLSAT Verbal Score'][i])
        except ValueError:
            #print("Oops!  That was no valid number.  Try again...")
            try:
                #print("value i is ",i,int(data['OLSAT Verbal Score'][i][0]+data['OLSAT Verbal Score'][i][1]))
                data['OLSAT Verbal Score'][i]= int(data['OLSAT Verbal Score'][i][0]+data['OLSAT Verbal Score'][i][1].copy(deep=True))
            except Exception as e:
                #print("value i is ",i,0)
                data['OLSAT Verbal Score'][i]=0
                #print("Oops!  That was no valid number.  Try again...")
        #print(int(data['OLSAT Verbal Score'][i]))
data.loc[data["OLSAT Verbal Score"]>30 , "OLSAT Verbal Score"] = 0.3*data['OLSAT Verbal Score']



#change percentages to out of 30
data.loc[data["OLSAT Verbal Score"]>30 , "OLSAT Verbal Score"] = 0.3*data['OLSAT Verbal Score']
print(data['OLSAT Verbal Score'])
    
#histogram of olsat verbal score
plt.hist(data['OLSAT Verbal Score'])
```
